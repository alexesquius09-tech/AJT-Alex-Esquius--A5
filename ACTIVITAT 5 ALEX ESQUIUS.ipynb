```python
class playlist():
    def __init__ (self, song_list):
        self.song_list= []
    def afegir_canço(self,song):
        self.song_list.append(song)
        return song
    
    def durada_total(self, total, durada):
        total=0
        for song in self.song_list:
            total += song.durada           
        return total
    
    def mostrar_playlist(self,):
        return self.song_list   
    
```
