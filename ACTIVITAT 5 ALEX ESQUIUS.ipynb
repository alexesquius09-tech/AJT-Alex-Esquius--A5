```python
class Playlist:
    def __init__(self,song_list):
        self.song_list= song_list
    def afegir_canco(self,song):
        self.song_list.append(song)
    def durada_total(self):
        total = 0
        for song in self.song_list:
            total += song.durada
        return total

    def mostrar_playlist(self):
        for song in self.song_list:
            print(song.titol)

    def canco_mes_llarga(self):
        canco_mes_llarga = self.song_list[0]
        for song in self.song_list:
            if song.durada > canco_mes_llarga.durada:
                canco_mes_llarga = song
        return canco_mes_llarga

        

    
```
